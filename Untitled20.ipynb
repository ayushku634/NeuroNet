```python
# =========================
# FULL PIPELINE: feed_to_gpt.csv → TE (no leakage) → patient agg → LGBM
# =========================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from category_encoders import TargetEncoder
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# =============== LOAD ===============
raw = pd.read_csv("feed_to_gpt.csv")

# Basic hygiene for known dtypes
for c in ['visit_start_dt','visit_end_dt','follow_up_dt','last_care_dt','next_care_dt','hot_spotter_identified_at']:
    if c in raw.columns:
        raw[c] = pd.to_datetime(raw[c], errors='coerce')

# Ensure booleans are int (safe for LightGBM)
for c in ['hot_spotter_readmission_flag','hot_spotter_chronic_flag','care_gap_flag']:
    if c in raw.columns and raw[c].dtype == bool:
        raw[c] = raw[c].astype(int)

# =============== PATIENT-LEVEL SPLIT (NO LEAKAGE) ===============
patients = raw[['patient_id','risk_score']].drop_duplicates()
train_pat, temp_pat = train_test_split(patients, test_size=0.30, random_state=42)
val_pat,   test_pat = train_test_split(temp_pat, test_size=0.50, random_state=42)

train_ids = set(train_pat['patient_id'])
val_ids   = set(val_pat['patient_id'])
test_ids  = set(test_pat['patient_id'])

# Helper: fit TE on training patients only, then transform all rows
def fit_transform_te_by_patients(df, col, target_col, train_id_set, smoothing=0.3):
    te = TargetEncoder(cols=[col], smoothing=smoothing)
    mask_tr = df['patient_id'].isin(train_id_set)
    # Fit on training rows
    df.loc[mask_tr, col + '_te'] = te.fit_transform(df.loc[mask_tr, col], df.loc[mask_tr, target_col]).values
    # Transform on non-training rows
    df.loc[~mask_tr, col + '_te'] = te.transform(df.loc[~mask_tr, col]).values
    return df, te

# =============== TARGET ENCODE (ROW-LEVEL) ===============
# Columns that were one-hot before → now target-encode
te_cols = [c for c in ['condition_name','msrmnt_type','msrmnt_sub_type','visit_type'] if c in raw.columns]

for col in te_cols:
    raw, _ = fit_transform_te_by_patients(raw, col, 'risk_score', train_ids, smoothing=0.3)

# =============== FEATURE ENGINEERING & AGGREGATION ===============

# ---- Patient base features (keep scaled versions; convert identified_at → day)
patient_base = (raw
    .groupby('patient_id', as_index=False)
    .agg({
        'risk_score':'first',
        'age':'first',
        'age_scaled':'first',
        'days_since_identified':'first',
        'days_since_identified_scaled':'first',
        'hot_spotter_readmission_flag':'max',
        'hot_spotter_chronic_flag':'max',
        'hot_spotter_identified_at':'first'
    })
)

if 'hot_spotter_identified_at' in patient_base.columns:
    # bogus 1900-01-01 → 0
    mask_1900 = patient_base['hot_spotter_identified_at'].dt.year == 1900
    patient_base.loc[mask_1900, 'hot_spotter_identified_at'] = pd.NaT
    patient_base['hot_spotter_identified_day'] = patient_base['hot_spotter_identified_at'].dt.day.fillna(0).astype(int)
    patient_base = patient_base.drop(columns=['hot_spotter_identified_at'])

# ---- Diagnosis agg
diag = raw[['patient_id','diagnosis_id','condition_type','condition_name','condition_name_te']].copy()
diagnosis_agg = (diag.groupby('patient_id', as_index=False)
                 .agg(num_diagnoses=('diagnosis_id', lambda s: int(np.isfinite(s).sum())),
                      num_chronic_conditions=('condition_type', lambda s: int((s == 'CHRONIC').sum())),
                      unique_conditions=('condition_name', lambda s: int(pd.Series(s).nunique(dropna=True))),
                      avg_condition_risk=('condition_name_te', 'mean'))
                 .fillna(0))

# ---- Care agg
care_cols = ['patient_id','care_id','days_since_last_care','last_care_dt','next_care_dt',
             'days_to_next_care','care_gap_flag','msrmnt_value',
             'msrmnt_type','msrmnt_sub_type','msrmnt_type_te','msrmnt_sub_type_te']
care = raw[[c for c in care_cols if c in raw.columns]].copy()

# If days_to_next_care missing, compute
if 'last_care_dt' in care.columns and 'next_care_dt' in care.columns:
    care['days_to_next_care'] = ((care['next_care_dt'] - care['last_care_dt']).dt.days
                                  if 'days_to_next_care' not in care.columns
                                  else care['days_to_next_care'])
care['days_to_next_care'] = care['days_to_next_care'].clip(lower=0, upper=365).fillna(0)

care_agg = (care.groupby('patient_id', as_index=False)
            .agg(num_care_records=('care_id', lambda s: int(np.isfinite(s).sum())),
                 avg_days_since_last_care=('days_since_last_care', 'mean'),
                 avg_days_to_next_care=('days_to_next_care', 'mean'),
                 has_care_gap=('care_gap_flag', 'max'),
                 avg_msrmnt_value=('msrmnt_value', 'mean'),
                 avg_msrmnt_type_risk=('msrmnt_type_te', 'mean'),
                 avg_msrmnt_subtype_risk=('msrmnt_sub_type_te', 'mean'),
                 unique_msrmnt_types=('msrmnt_type', lambda s: int(pd.Series(s).nunique(dropna=True))),
                 unique_msrmnt_subtypes=('msrmnt_sub_type', lambda s: int(pd.Series(s).nunique(dropna=True))))
            .fillna(0))

# ---- Visit agg
visit_cols = ['patient_id','visit_id','visit_type','visit_type_te','visit_start_dt','visit_end_dt',
              'follow_up_dt','readmsn_ind','visit_duration','days_to_follow_up']
visit = raw[[c for c in visit_cols if c in raw.columns]].copy()

# Compute durations and followup if needed
if 'visit_duration' not in visit.columns and {'visit_start_dt','visit_end_dt'}.issubset(visit.columns):
    visit['visit_duration'] = (visit['visit_end_dt'] - visit['visit_start_dt']).dt.days
visit['visit_duration'] = visit['visit_duration'].clip(lower=0).fillna(0)

if 'days_to_follow_up' not in visit.columns and {'follow_up_dt','visit_end_dt'}.issubset(visit.columns):
    visit['days_to_follow_up'] = (visit['follow_up_dt'] - visit['visit_end_dt']).dt.days
visit['days_to_follow_up'] = visit['days_to_follow_up'].clip(lower=0, upper=365).fillna(0)

visit['has_followup'] = visit['follow_up_dt'].notna().astype(int)

# Recency vs latest observed date
latest_end = visit['visit_end_dt'].max()
last_end = (visit.groupby('patient_id', as_index=False)['visit_end_dt'].max()
            .rename(columns={'visit_end_dt':'last_visit_end'}))
last_end['days_since_last_visit'] = (latest_end - last_end['last_visit_end']).dt.days
last_end = last_end[['patient_id','days_since_last_visit']]

# Readmission flag may be 't'/'f' or other; be robust
def count_true_t(s):
    s = s.astype(str).str.lower()
    return int((s == 't').sum())

visit_agg = (visit.groupby('patient_id', as_index=False)
             .agg(num_visits=('visit_id', lambda s: int(np.isfinite(s).sum())),
                  num_readmissions=('readmsn_ind', count_true_t),
                  unique_visit_types=('visit_type', lambda s: int(pd.Series(s).nunique(dropna=True))),
                  avg_visit_duration=('visit_duration', 'mean'),
                  avg_days_to_followup=('days_to_follow_up', 'mean'),
                  has_any_followup=('has_followup', 'max'),
                  avg_visit_type_risk=('visit_type_te', 'mean'))
             .merge(last_end, on='patient_id', how='left')
             .fillna(0))

# ---- Merge everything to patient level
merged = (patient_base
          .merge(diagnosis_agg, on='patient_id', how='left')
          .merge(care_agg,      on='patient_id', how='left')
          .merge(visit_agg,     on='patient_id', how='left')
          ).fillna(0)

# =============== URGENCY FEATURES (smaller days ⇒ larger score) ===============
urgency_cols = ['avg_days_since_last_care', 'avg_days_to_next_care', 'avg_days_to_followup', 'days_since_last_visit']
for col in urgency_cols:
    if col in merged.columns:
        merged[col + '_urgency'] = 1.0 / (1.0 + merged[col])

# Drop raw when a scaled exists
for col in list(merged.columns):
    if col.endswith('_scaled'):
        base = col[:-7]  # remove '_scaled'
        if base in merged.columns:
            merged.drop(columns=[base], inplace=True)

# Remove any pre-encoded columns from raw (avoid leakage/double use)
for c in [c for c in merged.columns if c.endswith('_encoded')]:
    merged.drop(columns=[c], inplace=True)


# Optional: save engineered table
# merged.to_csv("engineered_patients.csv", index=False)

# =============== BUILD MATRICES USING SAME PATIENT SPLIT ===============
def subset(df, ids):
    return df[df['patient_id'].isin(ids)].copy()

train_df = subset(merged, train_ids)
val_df   = subset(merged, val_ids)
test_df  = subset(merged, test_ids)

target_col = 'risk_score'
drop_cols = ['patient_id', target_col]
X_train, y_train = train_df.drop(columns=drop_cols, errors='ignore'), train_df[target_col]
X_val,   y_val   = val_df.drop(columns=drop_cols, errors='ignore'),   val_df[target_col]
X_test,  y_test  = test_df.drop(columns=drop_cols, errors='ignore'),  test_df[target_col]

print("Shapes:", X_train.shape, X_val.shape, X_test.shape)

# =============== TRAIN LGBM WITH EARLY STOPPING ===============
model = LGBMRegressor(
    objective='regression',
    boosting_type='gbdt',
    learning_rate=0.03,
    num_leaves=128,          # ⬅ bigger leaves
    max_depth=8,             # ⬅ allow deeper trees
    min_data_in_leaf=20,     # ⬅ smaller → more detailed splits
    feature_fraction=0.9,
    bagging_fraction=0.8,
    bagging_freq=5,
    n_estimators=4000,       # ⬅ more boosting
    random_state=42
)


model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    eval_metric='rmse'
)

# =============== EVALUATION ===============
y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred)
r2   = r2_score(y_test, y_pred)
print(f"\n✅ Test RMSE: {rmse:.4f}")
print(f"✅ Test R²:   {r2:.4f}")

# =============== FEATURE IMPORTANCE ===============
importances = model.feature_importances_
order = np.argsort(importances)[::-1][:30]
plt.figure(figsize=(9,10))
plt.barh(np.array(X_train.columns)[order], importances[order])
plt.gca().invert_yaxis()
plt.title("Top Feature Importances (Leakage-Free)")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
import torch
from torch import nn

# Scale numeric features
scaler = StandardScaler()
X_train_nn = scaler.fit_transform(X_train)
X_val_nn   = scaler.transform(X_val)
X_test_nn  = scaler.transform(X_test)

# Convert to tensors
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

X_train_t = torch.tensor(X_train_nn, dtype=torch.float32).to(device)
y_train_t = torch.tensor(y_train.values, dtype=torch.float32).view(-1,1).to(device)

X_val_t = torch.tensor(X_val_nn, dtype=torch.float32).to(device)
y_val_t = torch.tensor(y_val.values, dtype=torch.float32).view(-1,1).to(device)

X_test_t = torch.tensor(X_test_nn, dtype=torch.float32).to(device)
y_test_t = torch.tensor(y_test.values, dtype=torch.float32).view(-1,1).to(device)
```

```python
class RiskNet(nn.Module):
    def __init__(self, input_dim):
        super(RiskNet, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(256, 64),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(64, 1)
        )

    def forward(self, x):
        return self.net(x)

model_nn = RiskNet(X_train_t.shape[1]).to(device)
```

```python
optimizer = torch.optim.Adam(model_nn.parameters(), lr=1e-4)
loss_fn = nn.MSELoss()

best_val_loss = float('inf')
patience, patience_counter = 15, 0

for epoch in range(200):
    # Train
    model_nn.train()
    optimizer.zero_grad()
    pred = model_nn(X_train_t)
    loss = loss_fn(pred, y_train_t)
    loss.backward()
    optimizer.step()

    # Validate
    model_nn.eval()
    with torch.no_grad():
        val_pred = model_nn(X_val_t)
        val_loss = loss_fn(val_pred, y_val_t)

    print(f"Epoch {epoch+1:03d} | Train Loss: {loss.item():.4f} | Val Loss: {val_loss.item():.4f}")

    # Early stopping check
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_weights = model_nn.state_dict().copy()
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print("⏹ Early stopping triggered.")
            break

# Load best weights
model_nn.load_state_dict(best_weights)
```

```python
model_nn.eval()
with torch.no_grad():
    y_pred_nn = model_nn(X_test_t).cpu().numpy().flatten()

from sklearn.metrics import mean_squared_error, r2_score

rmse_nn = mean_squared_error(y_test, y_pred_nn)
r2_nn   = r2_score(y_test, y_pred_nn)

print("\n🤖 PyTorch Neural Network Results")
print(f"✅ RMSE: {rmse_nn:.4f}")
print(f"✅ R²:   {r2_nn:.4f}")
```

```python
import pandas as pd

# --- Step 0: Load all the cleaned CSVs ---
patient_df = pd.read_csv("patient_cleaned.csv")
diagnosis_df = pd.read_csv("diagnosis_cleaned.csv")
care_df = pd.read_csv("care_cleaned.csv")
risk_df = pd.read_csv("risk_cleaned.csv")
visit_df = pd.read_csv("visit_cleaned (1).csv")

# --- Step 1: Aggregate diagnosis data ---
diagnosis_agg = diagnosis_df.groupby('patient_id').agg(
    num_diagnoses=('diagnosis_id', 'count'),
    num_chronic_conditions=('condition_type', lambda x: (x == 'CHRONIC').sum()),
    unique_conditions=('condition_name', 'nunique'),
    has_cancer=('condition_name', lambda x: int('CANCER' in x.values)),
    has_diabetes=('condition_name', lambda x: int('DIABETES' in x.values)),
    has_hypertension=('condition_name', lambda x: int('HYPERTENSION' in x.values))
).reset_index()


# Convert date columns
care_df['last_care_dt'] = pd.to_datetime(care_df['last_care_dt'], errors='coerce')
care_df['next_care_dt'] = pd.to_datetime(care_df['next_care_dt'], errors='coerce')

# Fix unrealistic future/old placeholder dates
care_df.loc[care_df['next_care_dt'].dt.year > 2100, 'next_care_dt'] = np.nan
care_df.loc[care_df['next_care_dt'].dt.year < 1950, 'next_care_dt'] = np.nan

# Compute care gap durations
care_df['days_to_next_care'] = (care_df['next_care_dt'] - care_df['last_care_dt']).dt.days
care_df['days_to_next_care'] = care_df['days_to_next_care'].clip(0, 365).fillna(0)

# Aggregate care data
care_agg = care_df.groupby('patient_id').agg(
    num_care_records=('care_id', 'count'),
    avg_days_since_last_care=('days_since_last_care', 'mean'),
    avg_days_to_next_care=('days_to_next_care', 'mean'),
    has_care_gap=('care_gap_flag', 'max'),
    avg_msrmnt_value=('msrmnt_value', 'mean'),
    unique_msrmnt_types=('msrmnt_type_encoded', 'nunique'),
    unique_msrmnt_subtypes=('msrmnt_sub_type_encoded', 'nunique')
).reset_index()

care_agg.fillna(0, inplace=True)

# --- Step 3: Create visit-type counts (e.g., ER, Urgent Care, etc.) ---
visit_counts = (
    visit_df
    .groupby(['patient_id', 'visit_type'])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

# Rename columns (e.g., num_ER_visits, num_URGENT_CARE_visits)
visit_counts.columns = [
    f'num_{col.replace(" ", "_").upper()}_visits' if col != 'patient_id' else col
    for col in visit_counts.columns
]
# --- Process visit start/end dates ---
visit_df['visit_start_dt'] = pd.to_datetime(visit_df['visit_start_dt'], errors='coerce')
visit_df['visit_end_dt'] = pd.to_datetime(visit_df['visit_end_dt'], errors='coerce')

# Visit duration in days
visit_df['visit_duration_days'] = (visit_df['visit_end_dt'] - visit_df['visit_start_dt']).dt.days

# Time since last visit (relative to most recent visit date)
latest_date = visit_df['visit_end_dt'].max()
visit_last = visit_df.groupby('patient_id')['visit_end_dt'].max().reset_index()
visit_last['days_since_last_visit'] = (latest_date - visit_last['visit_end_dt']).dt.days

# Merge back to visit_df or include in aggregation later
# --- Clean and process follow_up_dt ---
visit_df['follow_up_dt'] = pd.to_datetime(visit_df['follow_up_dt'], errors='coerce')
visit_df['visit_end_dt'] = pd.to_datetime(visit_df['visit_end_dt'], errors='coerce')

# Ignore unrealistic future placeholder dates (e.g., year 8888)
visit_df.loc[visit_df['follow_up_dt'].dt.year > 2100, 'follow_up_dt'] = np.nan

# Days between visit end and scheduled follow-up
visit_df['days_to_followup'] = (visit_df['follow_up_dt'] - visit_df['visit_end_dt']).dt.days

# Whether a follow-up was scheduled
visit_df['has_followup'] = np.where(visit_df['follow_up_dt'].notna(), 1, 0)

# Cap extreme values and fill missing
visit_df['days_to_followup'] = visit_df['days_to_followup'].clip(0, 365).fillna(0)



visit_agg_basic = visit_df.groupby('patient_id').agg(
    num_visits=('visit_id', 'count'),
    num_readmissions=('readmsn_ind', lambda x: (x == 't').sum()),
    unique_visit_types=('visit_type', 'nunique'),
    avg_visit_duration=('visit_duration_days', 'mean'),
    max_visit_duration=('visit_duration_days', 'max'),
    avg_days_to_followup=('days_to_followup', 'mean'),
    has_any_followup=('has_followup', 'max')
).reset_index()

# Merge time-since-last-visit
visit_agg_basic = visit_agg_basic.merge(visit_last[['patient_id', 'days_since_last_visit']], on='patient_id', how='left')

# Merge visit-type counts with visit-level aggregates
visit_agg = visit_agg_basic.merge(visit_counts, on='patient_id', how='left')
visit_agg.fillna(0, inplace=True)

# --- Step 5: Merge everything together ---
merged_df = (
    patient_df
    .merge(diagnosis_agg, on='patient_id', how='left')
    .merge(care_agg, on='patient_id', how='left')
    .merge(visit_agg, on='patient_id', how='left')
    .merge(risk_df, on='patient_id', how='left')  # optional: only for training set
)

# --- Step 6: Fill missing values ---
merged_df.fillna({
    'num_diagnoses': 0,
    'num_chronic_conditions': 0,
    'unique_conditions': 0,
    'has_cancer': 0,
    'has_diabetes': 0,
    'has_hypertension': 0,
    'num_care_records': 0,
    'num_care_gaps': 0,
    'avg_msrmnt_value': 0,
    'num_visits': 0,
    'num_readmissions': 0,
    'unique_visit_types': 0
}, inplace=True)

# --- Step 7: Check results ---
print("✅ Merged dataset shape:", merged_df.shape)
print(merged_df.head())

# --- Step 8: Save for training ---
merged_df.to_csv("merged_training_data.csv", index=False)
print("✅ Final merged dataset saved as 'merged_training_data.csv'")
```
